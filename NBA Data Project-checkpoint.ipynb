```python
import pandas as pd
```

Load Data

```python
# Loading the NBA data
df = pd.read_csv('nba.csv')
```

```python
# Getting information about the null and non-null data
df.info()
```

```python
# no. of rows & Columns in the data
df.shape
```

```python
# Top 10 rows of the data 
df.head(10)
```

```python
# statistical info. about the numerical columns of data
df.describe().astype(int)
```

Data Cleaning

```python
# dropping null value rows of data
df.dropna(inplace=True)
```

```python
# Getting information about the null and non-null rows of data
df.info()
```

```python
# finding duplicate value rows in the data
df.duplicated()
```

```python
# removing duplicate data
df.drop_duplicates(inplace=True)
```

Data Transformation

```python
# adding a column named "BMI" to the data by the following formula
df['BMI'] = (df['Weight']/(70**2))*703
```

```python
# top 5 rows of data after adding new column
df.head()
```

EDA (Exploratory Data Analysis)

```python
# statistical measures of "Age','Weight' & 'Salary' Column
df[['Age','Weight','Salary']].describe().astype(int)
```

```python
# Average 'Age','Weight'&'Salary' of players according to their "Position"
avg_by_position = df.groupby('Position')[['Age','Weight','Salary']].mean().astype(int)
print(avg_by_position)
```

Visualization

```python
import matplotlib.pyplot as plt
```

```python
# Creating a histogram plot of players' age
plt.hist(df['Age'],bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title("Distribution of Players' Ages")
plt.show()
```

```python
# Box plot of players' salaries by positions
plt.figure(figsize=(10,6))
df.boxplot(column='Salary',by='Position',color='Orange')
plt.ylabel('Salary')
plt.title('Box plot of players Salaries by position')
plt.suptitle("")
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot for 'Age' vs 'Salary' by position in different colours
plt.figure(figsize=(10,6))
colors={'C':'Red','PF':'Blue','PG':'Grey','SF':'Green','SG':'Yellow'}
plt.scatter(df['Age'],df['Salary'],c=df['Position'].map(colors),alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title("Age vs Salary by Positions")
plt.legend(colors)
plt.show()
```

```python
# Top 10 players with highest salaries 
top_players = df.nlargest(10,'Salary')
top_players
```

```python
# Top 5 colleges with the most represented players
top_colleges = df['College'].value_counts().nlargest(5)
top_colleges
```

```python
# Pie Chart to show the distribution of players across different 'positions'
position_counts = df['Position'].value_counts()
plt.pie(position_counts,labels = position_counts.index,autopct='%1.1f%%',startangle=140,colors = ['Yellow','Green','Blue','Brown','Orange'])
plt.title('Position Distribution of Players')
plt.axis('equal')
plt.show()
```

```python
# Average salary of players by each team
teamwise_avg_salary = df.groupby('Team')['Salary'].mean().astype(int)
teamwise_avg_salary
```

```python
# Bar plot to visualize average salary of players by each team
plt.figure(figsize=(10,6))
teamwise_avg_salary.plot(kind='bar',color='Purple')
plt.xlabel('Team')
plt.ylabel('Salary')
plt.title('Teamwise average salary of players')
plt.xticks(rotation=90)
plt.show()
```
